```python
def calc_angle(long1, lat1, long2, lat2):
    dx = long2 - long1;
    dy = lat2 - lat1;
    angle = math.degrees(math.atan2(dy, dx));
    angle += 0 if angle >= 0 else 360
    return angle

def verify_angle(line):
    angle = int(line.split(",")[3])
    if (angle % 10 != 0):
        print("Angle not multiple of 10!")
    if (angle != float(line.split(",")[3])):
        print("Angle is float!")
    if not (0 <= angle < 360):
        print("Angle not between 0 and 350!")
def verify_move(line):
    prev_long, prev_lat = float(line[1]), float(line[2])
    angle = int(line[3])
    new_long, new_lat = float(line[4]), float(line[5])
    actual_angle = calc_angle(prev_long, prev_lat, new_long, new_lat)
    if abs(angle - actual_angle) > 0.000001:
        print(f"Invalid angle! Assumed {angle}, found {actual_angle}")
        print(line)
    dist = ((prev_long - new_long)**2 + (prev_lat-new_lat)**2)**0.5
    if abs(dist - 0.0003) > 0.000001:
        print(f"Did not move required distance! Moved {dist}")
    
def verify_start_end(start, end):
    start_long, start_lat = float(start[1]), float(start[2])
    end_long, end_lat = float(end[4]), float(end[5])
    dist = ((start_long - end_long)**2 + (start_lat-end_lat)**2)**0.5
    if dist >= 0.0003:
        print("Did not reach end start position! Distance", dist)
def verify_submission(lines):
    i = 0
    verify_start_end(lines[0].split(","), lines[-1].split(","))
    read_sensors = set()
    for line in lines:
        num = int(line.split(",")[0])
        if num != i + 1:
            print("Line number incorrect!")
        verify_angle(line)
        verify_move(line.split(","))
        read_sensors.add(line.split(",")[-1])
        i += 1
    if i >= 150:
        print("More than 149 lines!")
    if len(read_sensors) != 34: # for "null"
        print(f"Sensors off! Got {len(read_sensors)}")
```

```python
flightpaths = glob.glob("flightpath*")
strings = []
lengths = []
for flightpath in flightpaths:
    with open(flightpath, "r") as f:
        strings.append(f.readlines())
        lengths.append(len(strings[-1]))
```

```python
for i, submission in enumerate(strings):
    #print(flightpaths[i])
    verify_submission(submission)
```

```python
d = pd.DataFrame(data={"strings": strings, "lengths":lengths}, index=flightpaths)
```
